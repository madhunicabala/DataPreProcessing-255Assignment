[LP 1] Deal with the missing values in the credit analysis dataset. Explain what you did with the following columns:

1. Column A1 (at index 0)
2. Column A2 (at index 1)
3. Column A4 (at index 3)
4. Column A5 (at index 4)
5. Column A6 (at index 5)
6. Column A7 (at index 6)
7. Column A14 (at index 13)

**ANSWER:** Method used to deal with Missing values in the columns listed ie., numerically they are (0,1,3,4,5,6,13). For the ease, lets rename these columns to A1 to A14. The block[4] shows that what are the missing attributes in the crx.data dataset. There are two types of values in this dataset - Numerical and Categorical dataset.

Numerical dataset missing values is replaced using MEAN
whereas Categorical dataset missing values is replaced using MODE ie., frqeuently occuring value.

This is the STRATEGY adopted to fill the missing values and also verify, that there are no missing values for validation and ensuring whether there are no missing valuyes after the imputation.





```python
## Your code for LP 1 goes here.
missing_values = credit_approval.isnull().sum()
print("Columns with missing values:")
# Print the column names and their missing value counts
print(missing_values[missing_values > 0])
# Assign the column names A1 through A16 to the DataFrame
column_names = [f'A{i}' for i in range(1, 17)]
credit_approval.columns = column_names

# Display the columns with missing values using the new column names
missing_values = credit_approval.isnull().sum()
print("Columns with missing values:")
print(missing_values[missing_values > 0])

# Identify numerical and categorical columns with missing values
numerical_cols_with_missing = credit_approval.select_dtypes(include=np.number).columns[credit_approval.select_dtypes(include=np.number).isnull().any()]
categorical_cols_with_missing = credit_approval.select_dtypes(include='object').columns[credit_approval.select_dtypes(include='object').isnull().any()]

print("Numerical columns with missing values:", list(numerical_cols_with_missing))
print("Categorical columns with missing values:", list(categorical_cols_with_missing))

# Impute missing values in numerical columns with the mean
for col in numerical_cols_with_missing:
    mean_val = credit_approval[col].mean()
    credit_approval[col] = credit_approval[col].fillna(mean_val) # Modified line
    print(f"Filled missing values in '{col}' with the mean ({mean_val:.2f})")

# Impute missing values in categorical columns with the mode
for col in categorical_cols_with_missing:
    # Calculate mode, handle potential multiple modes by taking the first one
    mode_val = credit_approval[col].mode()[0]
    credit_approval[col] = credit_approval[col].fillna(mode_val) # Modified line
    print(f"Filled missing values in '{col}' with the mode ('{mode_val}')")

# Verify that there are no more missing values
print("\nMissing values after imputation:")
print(credit_approval.isnull().sum()[credit_approval.isnull().sum() > 0])
```


[LP 2] Use encoders and convert the categorical variables to numerical values: Columns A1, A4, A5, A6, A7, A9, A10, A12, A13






**ANSWER**

I have adopted "One-Hot encoding" technique inorder to convert the categorical values to numerical values. It is a type of encoder that converts categorical variables into a numerical format by creating new binary (0 or 1) columns for each unique category within the original column. This transformation makes the categorical data understandable to most machine learning algorithms, fulfilling the requirement to convert them to numerical values.


```python
## Your code for LP 2 goes here.
# Assign the column names A1 through A16 to the DataFrame
column_names = [f'A{i}' for i in range(1, 17)]
credit_approval.columns = column_names

# List of categorical columns to encode
categorical_cols_to_encode = ['A1', 'A4', 'A5', 'A6', 'A7', 'A9', 'A10', 'A12', 'A13']

# Perform one-hot encoding
credit_approval_encoded = pd.get_dummies(credit_approval, columns=categorical_cols_to_encode, drop_first=True)

# Display the first few rows of the encoded DataFrame and its shape
print("DataFrame after one-hot encoding:")
display(credit_approval_encoded.head())
print("\nShape of the DataFrame after encoding:", credit_approval_encoded.shape)
```

[HP 1] Describe your choices to preprocess these columns and explain why they are appropriate for the column.


**ANSWER**

Here is a breakdown of the preprocessing steps adopted and the reason behind those choices:

**Missing values handling**

1. Categorical Columns (A1, A4, A5, A6, A7) : Missing values in the categorical columns are imputd with Mode (the most frequent category). This preserves the distribution.

2. Numerical Columns (A2, A14) : Missing values in the numerical and misisng values can be imputed with Mean.

**Encoding and Scaling**

*   **Columns A1, A4, A5, A6, A7, A9, A10, A12, A13 (Categorical):** These categorical columns were transformed using **one-hot encoding**. One-hot encoding is appropriate because these columns are nominal. It creates binary (0 or 1) columns for each category.

While this increases dimensionality, it's necessary for most algorithms to process this type of data correctly.

*   **Columns A2, A3, A8, A11, A14, A15 (Numerical):** These numerical columns were subjected to **Standard Scaling**. Standard Scaling (Z-score normalization) transforms the data so that it has a mean of 0 and a standard deviation of 1. This is important for many machine learning algorithms because it prevents features with larger scales from dominating the learning process.
It ensures that all numerical features contribute relatively equally to the model.





[LP 3] Use an appropriate scaler to scale the numerical values to a suitable range of values. Briefly explain the process you used.

**ANSWER**

As written above, I used standard scaling technique to scale numerical values.

Columns A2, A3, A8, A11, A14, A15 (Numerical): These numerical columns were subjected to Standard Scaling. Standard Scaling (Z-score normalization) transforms the data so that it has a mean of 0 and a standard deviation of 1. This is important for many machine learning algorithms because it prevents features with larger scales from dominating the learning process.

It ensures that all numerical features contribute relatively equally to the model.

```python
## Your code for LP 3 goes here.
from sklearn.preprocessing import StandardScaler
import numpy as np # Import numpy if not already imported

# Identify the numerical columns in the encoded DataFrame
# We exclude the encoded categorical columns and the target variable (A16)
# Make sure to use the correct column names after renaming
numerical_cols = credit_approval_encoded.select_dtypes(include=np.number).columns.tolist()
if 'A16' in numerical_cols:
    numerical_cols.remove('A16') # Assuming A16 is the target variable and is not numerical after encoding

print("Numerical columns to scale:", numerical_cols)

# Initialize the StandardScaler
scaler = StandardScaler()

# Apply the scaler to the numerical columns
credit_approval_encoded[numerical_cols] = scaler.fit_transform(credit_approval_encoded[numerical_cols])

# Display the first few rows of the scaled DataFrame
print("\nDataFrame after scaling numerical features:")
display(credit_approval_encoded.head())
```
